# üöÄ VS Code + Colab T4 UI Launcher

Use this notebook when connected to Colab T4 via VS Code to launch the Wizard UI.

**Steps:**
1. Connect VS Code to Colab T4 (Ctrl+Shift+P ‚Üí Colab: Connect)
2. Run Cell 1 to check GPU
3. Run Cell 2 to upload ai-compiler folder
4. Run remaining cells to install and launch UI

```python
# 1Ô∏è‚É£ Check Environment & GPU
import subprocess
import os

print("üîç Checking GPU status...")
try:
    gpu_info = subprocess.check_output(['nvidia-smi', '--query-gpu=name', '--format=csv,noheader']).decode()
    print(f"‚úÖ GPU Detected: {gpu_info.strip()}")
except:
    print("‚ùå No GPU detected. Connect to Colab T4 first!")

print(f"\nüìÇ Current directory: {os.getcwd()}")
print(f"üìÅ Contents: {os.listdir('.')}")
```

```python
# 2Ô∏è‚É£ Upload ai-compiler Folder
# Run this cell, then upload the ai-compiler.zip file when prompted

import os

print("üì¶ Upload Options:")
print("="*50)
print("Option A: Upload ZIP file (click button below)")
print("Option B: Mount Google Drive (uncomment code below)")
print("="*50)

# Option A: Upload ZIP
try:
    from google.colab import files
    print("\nüì§ Click 'Choose Files' to upload ai-compiler.zip:")
    uploaded = files.upload()
    
    # Unzip
    for filename in uploaded.keys():
        if filename.endswith('.zip'):
            os.system(f'unzip -q -o {filename} -d /content/')
            print(f"‚úÖ Unzipped {filename}")
except:
    print("‚ö†Ô∏è Upload widget not available. Use Option B or manual upload.")

# Option B: Google Drive (uncomment these lines if you prefer Drive)
# from google.colab import drive
# drive.mount('/content/drive')
# !cp -r /content/drive/MyDrive/YOUR_PATH/ai-compiler /content/

# Check if successful
if os.path.exists('/content/ai-compiler'):
    os.chdir('/content/ai-compiler')
    print(f"\n‚úÖ Success! Now in: {os.getcwd()}")
else:
    print("\n‚ö†Ô∏è ai-compiler folder not found. Please upload it.")
```

```python
# 3Ô∏è‚É£ Install Dependencies
import os
import subprocess

# Make sure we're in ai-compiler
if not os.path.exists('pyproject.toml'):
    if os.path.exists('/content/ai-compiler'):
        os.chdir('/content/ai-compiler')
    else:
        print("‚ùå Run Cell 2 first to upload ai-compiler!")
        raise SystemExit()

print(f"üìÇ Working in: {os.getcwd()}")
print("üì¶ Installing dependencies... (this may take 1-2 minutes)")

subprocess.run(['pip', 'install', 'uv', '-q'], check=True)
subprocess.run(['uv', 'sync', '--extra', 'ui', '--extra', 'unsloth'], check=True)

print("\n‚úÖ Installation complete!")
```

```python
# 4Ô∏è‚É£ Verify Installation
import subprocess
subprocess.run(['uv', 'run', 'ai-compile', 'info'])
```

```python
# 5Ô∏è‚É£ Launch Wizard UI
import subprocess

print("üöÄ Launching Wizard UI...")
print("üîó Click the public link (gradio.live) below to open the app:")
print("="*50)

subprocess.run(['uv', 'run', 'ai-compile', 'ui2', '--share'])
```
