# üöÄ Finetune Compiler - Colab T4 Launcher

Launch the Wizard UI on Colab's Tesla T4 GPU.

**Steps:**
1. Connect to Colab T4 Runtime
2. Run Cell 1: Check GPU
3. Run Cell 2: Clone & Install
4. Run Cell 3: Launch UI

```
# 1Ô∏è‚É£ Check GPU
import subprocess

print("üîç Checking GPU...")
try:
    gpu = subprocess.check_output(['nvidia-smi', '--query-gpu=name', '--format=csv,noheader']).decode().strip()
    print(f"‚úÖ GPU: {gpu}")
    if 'T4' in gpu:
        print("üöÄ Tesla T4 detected - Unsloth will be used for fast training!")
except:
    print("‚ùå No GPU. Go to Runtime ‚Üí Change runtime type ‚Üí GPU")
```

```
# 2Ô∏è‚É£ Clone & Install
import os
import subprocess

REPO_URL = "https://github.com/khedhar108/Finetune-Compiler.git"
FOLDER = "Finetune-Compiler"

# Clone if not exists
if not os.path.exists(f'/content/{FOLDER}'):
    print(f"üì• Cloning {REPO_URL}...")
    subprocess.run(['git', 'clone', REPO_URL], cwd='/content')
else:
    print(f"üìÇ {FOLDER} already exists. Pulling latest...")
    subprocess.run(['git', 'pull'], cwd=f'/content/{FOLDER}')

os.chdir(f'/content/{FOLDER}')
print(f"üìÇ Working in: {os.getcwd()}")

# Install
print("\nüì¶ Installing dependencies...")
subprocess.run(['pip', 'install', 'uv', '-q'])
subprocess.run(['uv', 'sync', '--extra', 'ui', '--extra', 'unsloth'])

print("\n‚úÖ Ready!")
```

```
# 3Ô∏è‚É£ Verify Installation
import subprocess
subprocess.run(['uv', 'run', 'ai-compile', 'info'])
```

```
# 4Ô∏è‚É£ Launch Wizard UI
import subprocess

print("üöÄ Starting Wizard UI...")
print("üîó Click the gradio.live link below:")
print("=" * 50)

subprocess.run(['uv', 'run', 'ai-compile', 'ui2', '--share'])
```
